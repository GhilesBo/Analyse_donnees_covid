```python
multi_index = df.set_index(['continent' , 'location'])
```

```python
country_sum = multi_index.groupby(['location']).agg({'total_cases':['sum'], 'total_deaths':['sum'] })
```

```python
country_sum.columns = ['nb_cas_total' , 'nb_total_morts']
```

```python
country_sum
```

tri de nb_cas_total et nb_total_morts

```python
sorted_df_cases = country_sum.sort_values(['nb_cas_total'], ascending=True)
sorted_df_deaths = country_sum.sort_values(['nb_total_morts'], ascending=False)
```

```python
sorted_df_cases.head(29)
```

```python
sort_filtered_cases = sorted_df_cases.drop(['World', 'Upper middle income', 'High income', 'Europe', 'North America', 'South America', 'Asia', 'Lower middle income', 'European Union'])
```

```python
sort_filtered_deaths = sorted_df_deaths.drop(['World', 'Upper middle income', 'High income', 'Europe', 'North America', 'South America', 'Asia', 'Lower middle income', 'European Union'])
```

```python
sort_filtered_cases.head(10)
```

```python
sort_filtered_deaths.head(10)
```

```python
sort_filtered_cases['nb_cas_total'].plot(kind='pie', figsize=(6, 6))
```

```python
sort_filtered_deaths['nb_total_morts'].plot(kind='pie', figsize=(6, 6))
```
